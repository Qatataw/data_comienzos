```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
pd.options.display.max_rows = 100

df = pd.read_csv(r'C:\Users\Eva\Desktop\Python\programas\data\data_sheets\MrBeast_data.csv')
df = df.set_index('id')

df = df[['title', 'description', 'publishTime', 'duration_seconds', 'viewCount',
       'likeCount', 'commentCount']]
```

```python
df_view = df.query('viewCount > 1000000')
df = df.loc[~df['viewCount'].isna()]
```

```python
df['viewCount'] = df['viewCount'].astype('int')
df['likeCount'] = df['likeCount'].astype('int')
df['publishTime'] = pd.to_datetime(df['publishTime'])
```

```python
df['like_to_view'] = df['likeCount'] / df['viewCount']
```

```python
df_to = df.tail(1)
df_to = pd.concat([df,df_to])
```
